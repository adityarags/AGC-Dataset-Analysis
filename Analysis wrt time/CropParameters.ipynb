```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
import datetime
from sklearn.impute import SimpleImputer
```

```python
os.chdir("/content/Dataset")
```

```python
files = os.listdir()
files.remove("Economics.pdf")
files.remove("ReadMe.pdf")
files.remove("Weather")
```

```python
files
```

```python
CropParameters = {}
for i in files:
  filepath = os.path.join(i, "CropParameters.csv")
  CropParameters[i] = pd.read_csv(filepath)
```

```python
CropParameters[files[0]].head()
```

```python
si = SimpleImputer(strategy = "mean")
for i in files:
  column_names = CropParameters[i].columns
  values = CropParameters[i].values
  df = si.fit_transform(values)
  CropParameters[i] = pd.DataFrame(df, columns = column_names)
```

```python
for i in range(len(files)):
  CropParameters[files[i]]['%Time'] = pd.to_datetime(CropParameters[files[i]]['%Time'], unit='D', origin='1899-12-30')
```

```python
for i in range(len(files)):
  plt.plot(CropParameters[files[i]]["%Time"], CropParameters[files[i]]["Stem_elong"])
plt.legend(files)
plt.title("Stem_elong vs Time")
plt.xlabel("Time")
plt.ylabel("Stem_elong")
plt.show()
```

```python
for i in range(len(files)):
  plt.plot(CropParameters[files[i]]["%Time"], CropParameters[files[i]]["Stem_thick"])
plt.legend(files)
plt.title("Stem_thick vs Time")
plt.xlabel("Time")
plt.ylabel("Stem_thick")
plt.show()
```

```python
for i in range(len(files)):
  plt.plot(CropParameters[files[i]]["%Time"], CropParameters[files[i]]["Cum_trusses"])
plt.legend(files)
plt.title("Cum_trusses vs Time")
plt.xlabel("Time")
plt.ylabel("Cum_trusses")
plt.show()
```

```python
for i in range(len(files)):
  plt.plot(CropParameters[files[i]]["%Time"], CropParameters[files[i]]["stem_dens "])
plt.legend(files)
plt.title("stem_dens vs Time")
plt.xlabel("Time")
plt.ylabel("stem_dens")
plt.show()
```
