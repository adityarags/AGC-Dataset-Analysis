```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
import datetime
from sklearn.impute import SimpleImputer
```

```python
os.chdir("/content/Dataset")
```

```python
files = os.listdir()
files.remove("Economics.pdf")
files.remove("ReadMe.pdf")
files.remove("Weather")
```

```python
files
```

```python
GrodanSens = {}
for i in files:
  filepath = os.path.join(i, "GrodanSens.csv")
  GrodanSens[i] = pd.read_csv(filepath)
```

```python
GrodanSens[files[0]].describe()
```

```python
GrodanSens[files[0]].head()
```

```python
si = SimpleImputer(strategy = "mean")
for i in files:
  column_names = GrodanSens[i].columns
  values = GrodanSens[i].values
  df = si.fit_transform(values)
  GrodanSens[i] = pd.DataFrame(df, columns = column_names)
```

```python
GrodanSens[files[0]].head()
```

```python
GrodanSens[files[1]].describe()
```

```python
for i in range(len(files)):
  GrodanSens[files[i]]['%time'] = pd.to_datetime(GrodanSens[files[i]]['%time'], unit='D', origin='1899-12-30')
```

```python
np.unique(GrodanSens[files[0]]['%time'].shape)
```

```python
parameters = list(GrodanSens[files[0]].columns)
parameters.remove("%time")
parameters
```

```python
plt.figure(figsize = (20, 10))
param = "EC_slab1"
for i in range(len(files)):
  plt.plot(GrodanSens[files[i]]["%time"], GrodanSens[files[i]][param])
plt.legend(files, loc = "best")
plt.title(f"{param} vs Time")
plt.xlabel("Time")
plt.ylabel(f"{param}")
plt.show()
```
