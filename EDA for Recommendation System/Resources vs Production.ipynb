```python
import pickle
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.impute import SimpleImputer
```

## AICU

```python
Productiondf = pd.read_csv("/content/drive/MyDrive/Dataset/AICU/Production.csv")
```

```python
Productiondf.head()
```

```python
Productiondf.info()
```

```python
np.unique(Productiondf["%time"])
```

```python
si = SimpleImputer(strategy = "mean")

column_names = Productiondf.columns
values = Productiondf.values
df = si.fit_transform(values)
Productiondf = pd.DataFrame(df, columns = column_names)
```

```python
Productiondf["%time"] = pd.to_datetime(Productiondf["%time"], unit='D', origin='1899-12-30')
```

```python
Productiondf.head()
```

```python
Resourcedf = pd.read_csv("/content/drive/MyDrive/Dataset/AICU/Resources.csv")
```

```python
Resourcedf.head()
```

```python
np.unique(Resourcedf["%Time "])
```

```python
Resourcedf["%Time "] = pd.to_datetime(Resourcedf["%Time "], unit='D', origin='1899-12-30')
```

```python
Resourcedf.head()
```

```python
np.unique(Productiondf["%time"])
```

```python
Resourcedf.rename(columns={"%Time ": "Time"}, inplace=True)
```

```python
cols = list(Resourcedf.columns)[1:]
additionalColumns = {column: [] for column in cols}

for x in np.unique(Productiondf["%time"]):
  for ind, x in enumerate(Resourcedf[(Resourcedf['Time'] <= x)].mean()):
    additionalColumns[cols[ind]].append(x)

for k, v in additionalColumns.items():
  Productiondf[k] = v
```

```python
Productiondf.head()
```

```python
corr_matrix = Productiondf.corr()
plt.figure(figsize = (10, 10))
sns.heatmap(corr_matrix, annot = True)
plt.show()
```
