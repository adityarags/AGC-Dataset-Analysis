```python
import pickle
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.impute import SimpleImputer
```

## AICU

```python
TQualitydf = pd.read_csv("/content/drive/MyDrive/Dataset/AICU/TomQuality.csv")
```

```python
TQualitydf.head()
```

```python
TQualitydf.info()
```

```python
np.unique(TQualitydf["%time"])
```

```python
si = SimpleImputer(strategy = "mean")

column_names = TQualitydf.columns
values = TQualitydf.values
df = si.fit_transform(values)
TQualitydf = pd.DataFrame(df, columns = column_names)
```

```python
TQualitydf["%time"] = pd.to_datetime(TQualitydf["%time"], unit='D', origin='1899-12-30')
```

```python
TQualitydf.head()
```

```python
Resourcedf = pd.read_csv("/content/drive/MyDrive/Dataset/AICU/Resources.csv")
```

```python
Resourcedf.head()
```

```python
np.unique(Resourcedf["%Time "])
```

```python
Resourcedf["%Time "] = pd.to_datetime(Resourcedf["%Time "], unit='D', origin='1899-12-30')
```

```python
Resourcedf.head()
```

```python
np.unique(TQualitydf["%time"])
```

```python
Resourcedf.rename(columns={"%Time ": "Time"}, inplace=True)
```

```python
cols = list(Resourcedf.columns)[1:]
additionalColumns = {column: [] for column in cols}

for x in np.unique(TQualitydf["%time"]):
  for ind, x in enumerate(Resourcedf[(Resourcedf['Time'] <= x)].mean()):
    additionalColumns[cols[ind]].append(x)

for k, v in additionalColumns.items():
  TQualitydf[k] = v
```

```python
TQualitydf.head()
```

```python
corr_matrix = TQualitydf.corr()
plt.figure(figsize = (10, 10))
sns.heatmap(corr_matrix, annot = True)
plt.show()
```
