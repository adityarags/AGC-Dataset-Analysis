```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
import seaborn as sns
import pickle
```

```python
columns = ["CropType", "CropDays", "SoilMoisture", "Temperature", "Humidity", "NeedIrrigation"]
df = pd.read_csv("/content/Agriculture-Automation-/Dataset/datasets.csv", names = columns)
df.head()
```

```python
print("Shape of the dataset:",df.shape)
df.info()
```

No null values

```python
df.describe()
```

```python
countvals = []
crops = ["Wheat", "Ground Nuts", "Garden flowers", "Maize", "Paddy", "Potato", "Pulse", "SugerCane", "Coffee"]
plt.figure(figsize = (20, 5))
for i in np.unique(df["CropType"]):
  countvals.append(df["CropType"].value_counts()[i])

plt.bar(crops, countvals)
plt.show()
```

```python
df.hist(figsize = (10, 10))
plt.show()
```

```python
plt.plot(df["Temperature"], ".")
plt.show()
```

Outlier in Temperature

```python
df.drop(df[(df['Temperature'] > 100)].index, inplace=True)
```

```python
plt.plot(df["Temperature"], ".")
plt.show()
```

Outlier Removed!
